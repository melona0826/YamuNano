Linear Regression 수식을 이용하여 노이즈가 포함된 데이터를 이용한 Linear Regression을 확인해보아요!

```python
# 필요한 라이브러리들 import
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# 파라미터 설명


## slope : 데이터 직선의 기울기
## bias : 데이터 직선의 y절편
## dataRange : 데이터의 x축 범위 (0~dataRange)
## dataNumber : 데어티의 개수


## randomNoizeFactor : 노이즈 정도 (값이 클 수록 노이즈가 커짐)

```python
slope = 2
bias = 2
dataRange = 300
dataNumber = 50

randomNoizeFactor = 50.0
```


$W = (X^TX)^{-1}X^TY$

```python
x = np.linspace(0, dataRange, dataNumber)

y = x * slope + bias + np.random.normal(0, randomNoizeFactor, dataNumber)
y = np.reshape(y, (y.shape[0], 1))

x = np.reshape(x, (x.shape[0], 1))
x = np.append(np.ones((x.shape[0], 1)), x, axis=1)

w = np.linalg.inv(x.T @ x) @ x.T @ y

print("계산된 W")
print(w)
```

## 노이즈가 포함된 여러개의 데이터 확인

```python
plt.axis([0, dataRange + 10, 0, dataRange*slope + bias + 50])
plt.scatter(x[:,1], y[:])

plt.show()
```

# 계산한 $W$를 이용한 직선 확인하기


```python
x_plot = np.linspace(0, dataRange, dataNumber)
y_plot = w[0] + w[1] * x_plot

plt.axis([0, dataRange + 10, 0, dataRange*slope + bias + 50])

plt.scatter(x[:,1], y[:])

plt.plot(x_plot, y_plot, 'r')
plt.show()
```
