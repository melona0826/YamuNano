2개의 점을 Linear Regression 수식을 이용해 실제로 직선을 잘 구하는지 확인해보아요!

```python
# 필요한 라이브러리들 import
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

$X = \begin{bmatrix}1&2 \\ 1&7 \end{bmatrix}$

$Y = \begin{bmatrix}5 \\ 10 \end {bmatrix}$

$W = (X^TX)^{-1}X^TY$

```python
x = np.array([[1, 2], [1, 7]])
y = np.array([5, 10])

w = np.linalg.inv(x.T @ x) @ x.T @ y

print(w)
```

## X의 두 점 확인하기

```python
plt.axis([0, 10, 0, 15])
plt.scatter(x[0][1], y[0])
plt.scatter(x[1][1], y[1])
plt.show()
```

# 계산한 $W$를 이용한 직선 확인하기


```python
x_plot = np.arange(0, 10)
y_plot = w[0] + w[1] * x_plot

plt.axis([0, 10, 0, 15])

plt.scatter(x[0][1], y[0])
plt.scatter(x[1][1], y[1])

plt.plot(x_plot, y_plot)
plt.show()
```
